## Task 1 - Linear Regression
We will use the craigslist housing dataset to perform linear regression to find the model parameters. After fitting the model we will predict housing prices from the test sizes. To check the performance we will calculate the root mean squared error. Please follow the notebook instructions step by step.
### 1.1 Import the libaries.

```python
# Import the required libraries (numpy, pandas, matplotlib.pyplot) and uncommment the prepared imports below.
# --- Code goes here ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

### 1.2 Load the data into a pandas dataframe.
Find The Pandas read_csv() function documentation here (For today you are fine without the documentation)

https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html


```python
# Load housing dataset into pandas dataframe.
# Use commands df=pd.read_csv('./housing.csv',sep=',').
df = pd.read_csv('./housing.csv',sep=',')
# Add a new column df['sqm']=0.09290304*df['sqft'] to calculate size in square meter.
df['sqm']=0.09290304*df['sqft']
# Inspect dataframe using df.head()
df.head()
# --- Code goes here ---
```

### 1.3 Plot the dataset.
Find the matplotlib.pyplot documentation for scatterplots and lineplots here

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# Plot the data via matplotlib: Use scatterplot.
# --- Code goes here ---
plt.scatter(df['sqm'], df['price'])
plt.xlabel("Price")
plt.ylabel("$m^2$")
plt.title("price based on $m^2$")

#plt.xaxis.set_tick_params(width = 5)
#plt.xaxis.set_tick_params(lenght = 10)
#plt.set_xticks(ticks=np.arrange(0,350,50))

plt.legend()
plt.show()
```

### 1.4 Perform train test split.
Train test splits usually take numpy arrays as input. We can read a pandas series by using "values" such that arrayname=df['colname'].values, for more info refer to 

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.values.html

Train test splits usually prefer 2d-array data input shapes. If reshape is required refer to 

https://numpy.org/doc/stable/reference/generated/numpy.reshape.html

```python
# Get X and y form pandas dataframe (reshape data into matrix format using array.reshape(-1, 1)).
X = df['sqm'].to_numpy()
y = df['price'].to_numpy()
# --- Code goes here ---
print(f"X:{X}")
print(f"y:{y}")
```

Find the sklearn documentation for train test split here.

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
# Perform a Train Test Split.
# --- Code goes here ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75, shuffle = False)
```

### 1.5 Fit the training data to a linear regression model and predict the the output for the test input data.
Find the sklearn documentation for linear regression here:

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
# Create linear regression object.
# --- Code goes here ---
reg = LinearRegression()
reg.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))
```

```python
# Train the model using the training sets.
# --- Code goes here ---
LinearRegression()
print(f"\n{reg.coef_}\n")
```

```python
# Make predictions using the testing set.
# --- Code goes here ---
y_pred = reg.predict(X_test.reshape(-1, 1))
```

```python
# Print out the parameters
print(f"Coeff: {reg.coef_}, {reg.intercept_}")
print(f"Parameters: {y_pred}")
```

### 1.6 Calculate the root mean squared error to check the performance.
Find the sklearn documentation for mean squared error calculation here:
    
https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html#sklearn.metrics.mean_squared_error

```python
# Calculate the root mean squared error by using the sklearn mean squared error function
# --- Code goes here ---
error = np.sqrt(mean_squared_error(y_test, y_pred))
error
```

```python
# Plot prediction outputs using a scatterplot for the test data and a line plot for the predictions.
# --- Code goes here ---
plt.scatter(df['sqm'], df['price'])
plt.plot(X_test, y_pred, 'red')
plt.xlabel("Price")
plt.ylabel("m^2")
plt.title("m^2 based on price")
plt.legend()
plt.show()
```

## Task 3 - Classification 
### 3.1 Perform train test split with 25% test data


```python
X = df[['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm']].values
y = df['class'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle = False)
```

### 3.2 Train with K-Nearest Neighbor Classifier (Set K=3)
Find the sklearn documentation for K-Nearest Neighbors here

https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

```python
# Train the model
neigh = KNeighborsClassifier(n_neighbors=3)
#neigh.fit(X_train.reshape(-1, 1), y_train)
neigh.fit(X_train, y_train)
```

### 3.3 Predict class labels for test data

```python
# --- Predict class labels
#y_pred = neigh.predict(X_test.reshape(-1, 1))
y_pred = neigh.predict(X_test)
```

### 3.4 Calculate prediction accuracy score
Find the sklearn documentation here

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html

https://scikit-learn.org/stable/modules/model_evaluation.html#accuracy-score


```python
# --- Determine accuracy
accuracy_score(y_test, y_pred)
```

### 3.5 Use a decision tree to classify the iris dataset and compare the accuracy to the K-Nearest Neighbor Method. 
Find the sklearn documentation here 

https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

```python
clf = DecisionTreeClassifier(max_depth = 2)
clf.fit(X_train, y_train)
y_pred_tree = clf.predict(X_test)
accuracy_score(y_test, y_pred_tree)
```
